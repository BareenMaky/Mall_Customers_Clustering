```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

#Import cvs file
df = pd.read_csv('Mall_Customers.csv')



#To show first 5 Rows in data
df.head()

#To get number of Rows and Columns
print(df.shape)

#To get information about data such as null values and datatypes
df.info()

#Display the data before dropping the 'CustomerID' column
print("Data before dropping 'CustomerID':")
print(df.head())

# Dropping the 'CustomerID' column
df.drop('CustomerID', axis=1, inplace=True)

# Displaying the data after dropping the 'CustomerID' column
print("\nData after dropping 'CustomerID':")
print(df.head())


#To get sum of null values in each column
df.isnull().sum()

#to show minimum age of customers
min_age = df['Age'].min()
print(min_age)

#to show maximum age of customers
max_age = df['Age'].max()
print(max_age)

#to show age of customers
ages = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
counts = [100, 350, 450, 300, 200, 150, 100]

plt.bar(ages, counts, color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.title('Customer Age Distribution')
plt.show()

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
Annual_Income = [75, 80, 60, 90, 85]
spending_scores = [60, 65, 55, 70, 75]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot for Annual Income
ax[0].bar(categories,Annual_Income , color='skyblue')
ax[0].set_title('Annual Income')
ax[0].set_xlabel('Categories')
ax[0].set_ylabel('Scores')

# Plot for spending scores
ax[1].bar(categories, spending_scores, color='lightgreen')
ax[1].set_title('Spending Scores')
ax[1].set_xlabel('Categories')
ax[1].set_ylabel('Scores')

plt.show()


# Creating a boxplot for the selected columns(Annual Income and Spending Score)
selected_columns = ['Annual Income (k$)', 'Spending Score (1-100)']

plt.figure(figsize=(8, 6))
df[selected_columns].boxplot()
plt.title('Boxplot of Annual Income and Spending Score')
plt.ylabel('Value')
plt.xticks(ticks=range(1, len(selected_columns) + 1), labels=selected_columns)
plt.show()


# Creating a boxplot for the selected columns, removing outliers from 'Annual Income'
selected_columns = ['Annual Income (k$)', 'Spending Score (1-100)']

plt.figure(figsize=(8, 6))
df[selected_columns].boxplot(showfliers=False)  # Set showfliers to False to remove outliers
plt.title('Boxplot of Annual Income and Spending Score (without outliers in Income)')
plt.ylabel('Value')
plt.xticks(ticks=range(1, len(selected_columns) + 1), labels=selected_columns)
plt.show()


# Creating a boxplot for the selected columns (Age and Gender)
selected_columns = ['Age', 'Gender']

plt.figure(figsize=(8, 6))
df[selected_columns].boxplot(by='Gender')
plt.title('Boxplot of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()


#Store 3rd and 4th column values in variable X
X = df.iloc[:,[2,3]].values
print(X)

#Elbow method
#optimum number of clusters will be 5, and Kmeans algorithm will be run 10 times
wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)

# Within Clusters Sum of Squares
print(wcss)

#elbow method graph
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
#the output shows that the number of optimal clusters = 5


kmeans = KMeans(n_clusters=5, init='k-means++', n_init=10, random_state=0)

# return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)

print(Y)
#Ouutput is an array representing the cluster assigned for each data point


plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# plot the center points(centroids)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
